## Ejercicio 5
Realiza un conversor de grados Celsius a Fahrenheit. El conversor tiene que recibir el input del usuario y devolver la conversión.

La conversión de grados sigue la siguiente fórmula: F = (C × 9 / 5) + 32 

```python
celsius = int(input("Por favor, introduzca los °C ")) 
''' Usuario introduce los grados celsius que quiere transformar,
y lo transformamos en un integer, ya que no se puede operar con strings'''
fahrenheit = (celsius * 9 / 5) + 32 # Aplicamos la formula que transforma celsius en farenheit

print(celsius, "grados celsius equivalen a", fahrenheit, "grados fahrenheit") # imprimimos en pantalla la variable fahrenheit
```
