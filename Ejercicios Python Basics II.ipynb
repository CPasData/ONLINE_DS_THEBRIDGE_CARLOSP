## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

````python
import math
grados = int(input("Por favor, ingresa los grados (°)"))
radianes = grados * (math.pi/180)

print(radianes)
````

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

````python
base = int(input("Introduzca la base en centimetros"))
altura = int(input("Introduzca la altura del paralelogramo en centimetros"))

area = base * altura
print(f"El area de tu paralelogramo son {area} centimetros")
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10

## Ejercicio 10
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."

1. Pásalo todo a mayúsculas
2. Pásalo todo a minúsculas
3. Solo la primera letra de cada palabra en mayúscula, el resto en minúscula
4. Crea una lista dividiéndolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

````python
mi_string = "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."
mi_lista = mi_string.split()


print(mi_string.upper())
print(mi_string.lower())
print(mi_string.capitalize())
print(mi_lista)
print(mi_string.replace(",",";"))
print(mi_string.replace("a", ""))
````
