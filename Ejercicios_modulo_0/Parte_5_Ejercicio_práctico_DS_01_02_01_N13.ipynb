## Ejercicio 12
Solicitar al usuario que ingrese su dirección email. Imprimir un mensaje indicando si la dirección es válida o no, valiéndose de una función para decidirlo. Una dirección se considerará válida si contiene el símbolo "@".

`````python
resultado = input("Por favor, ingrese su cuenta de email: ")

def verificar_email(resultado):
    for elemento in resultado:
        if elemento == "@":
            return "¡Enhorabuena, el registro se ha completado!"
        else:
            continue
    return "Dirección de email no válida"

print(verificar_email(resultado))
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
DNI = 55555555
palabra ='TRWAGMYFPDXBNJZSQVHLCKE'
print('La letra del DNI es: ', palabra[DNI%23])
`````

`````python
DNI = input("Ingrese su DNI, por favor")

def verificar_DNI(DNI):
    palabra ='TRWAGMYFPDXBNJZSQVHLCKE'
# Verificar la longitud del input. Excluimos los menores de 7 y mayores de 9 (contando con la letra)
    if len(DNI) < 7 or len(DNI) > 9:
        return "Lo siento, el numero de digitos es incorrecto"
# verificar la letra    
    letra = DNI[-1].upper()
    numero = int(DNI[:-1])
    letra_correcta = palabra[numero%23]
    if letra != letra_correcta:
        return True
    else:
        return "Lo siento, la letra no es correcta"

print(verificar_DNI(DNI))
    
    
`````
